```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv(r"https://raw.githubusercontent.com/JohnMHoyt24/brazil-wildfires/main/amazon.csv",encoding='latin1')
df.head(5)
```

```
df.dtypes #This is how I can check the data types for the columns of the dataframe
```

```
df.number.describe().round(2)
#This provides a statistical summary pertaining to the number of fires
```

```
df.corr().round(2)
#Judging by the correlation coefficients provided, there's a moderate positive relationship between the year and the number of fires
```

```
fire_corr = df.corr()
fig, ax = plt.subplots(figsize = (12,8))
sns.heatmap(fire_corr, annot=True)
```

```
plt.figure(figsize=(48,16)) #figsize= (x,y)
plt.bar(df['state'], df['number'])
plt.xlabel('State')
plt.ylabel('# of Fires')
```
